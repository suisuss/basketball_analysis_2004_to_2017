# The Overall Proportion between USA Players and International Players

#### read in csv file
#### 3 csv files: 1) demographics.csv 2) stats.csv 3) salaries.csv


```python
demographics_path = "../data/demographics.csv"
demographics = pd.read_csv(demographics_path)
demographics.head()
```

```python
stats_path = "../data/stats.csv"
stats = pd.read_csv(stats_path)
stats.head()
```

```python
salary_path = "../data/salaries.csv"
salary_df = pd.read_csv(salary_path)
salary_df.head()
```

#### choose the columns I need and clean data

```python
demographics_clean = demographics.drop(columns=["collage","born","height","weight"]).dropna(how="any")
demographics_clean.head()
```

```python
stats_clean = stats[["year","player"]]
stats_clean.head()
```

```python
salary_df.dropna(how='any',inplace=True)
salary_df
```

#### merge these dataframes

```python
merge_df = pd.merge(demographics_clean,stats_clean,how="right",left_on="name",right_on="player")
merge_df = merge_df.dropna(how='any').drop(columns='player')
merge_df.head()
```

```python
merge_salary_df = pd.merge(merge_df,salary_df,how='left', left_on=['name','year'],right_on=['name','season_start'])
merge_salary_df.head()
```

#### choose the columns I need and clean data

```python
merge_salary_df.drop(columns=["register_value","season_start","season_end","full_team_name"],inplace=True)
merge_salary_df.dropna(how='any',inplace=True)
merge_salary_df.head()
```

#### find out whether the given cities and states are in USA

```python
url = "https://maps.googleapis.com/maps/api/geocode/json?"
country_list=[]
Lat_list = []
Lng_list = []
for index,row in merge_salary_df.iterrows():
    address = f"{row['birth_city']},{row['birth_state']}"
    params = {"address":address,
             "key":gkey}
    response = requests.get(url,params=params).json()

    try:
        if "USA" in response['results'][0]['formatted_address']:
            country_list.append('USA')
            
        else:
            country_list.append('International')
        Lat_list.append(response['results'][0]['geometry']['location']['lat'])
        Lng_list.append(response['results'][0]['geometry']['location']['lng'])
    except:
        print(f'{address} is not found.')
```

```python
merge_salary_df["International Player/USA Player"]=country_list
merge_salary_df["Lat"] = Lat_list
merge_salary_df["Lng"] = Lng_list
```

```python
merge_salary_df
```

#### The Overall Proportion between USA Players and International Players
#### get the non-duplicated names dataframe

```python
unique_name_df = merge_salary_df.drop_duplicates(subset='name')
```

#### get the numbers needed to plot

```python
unique_name_players_number = unique_name_df.groupby('International Player/USA Player').count()
unique_name_players_number
```

```python
number_usa = unique_name_players_number.loc['USA','name']
number_international = unique_name_players_number.loc['International','name']
```

#### plotting

```python
#set up the params
labels = ['International Player','USA Player']
size = [number_international,number_usa]
explode = (0.1, 0)
colors = ['#CE1141','#00538C']

patches, texts, autotexts =plt.pie(size, colors = colors, labels=labels, autopct='%1.1f%%', startangle=170,pctdistance=0.85)
for text in texts:
    text.set_color('#002B5E')
for autotext in autotexts:
    autotext.set_color('white')
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')  
plt.tight_layout()


plt.title("The Overall Proportion of NBA Player of the Years from 2007 - 2017 ",fontsize = 15)
plt.legend(loc='lower right')
print("")
```

# The Trend of NBA Player Composition of the Years from 2007 - 2017

### group the data by multiple columns 'year' and 'International Player/USA Player'

```python
internatioanlplayers_and_usaplayers_each_year = merge_salary_df.groupby(['year','International Player/USA Player']).count()
total_players_each_year = merge_salary_df.groupby('year').count()
```

```python
internatioanlplayers_and_usaplayers_each_year
```

```python
total_players_each_year
```

```python
#set up the params 
percentage_each_year = internatioanlplayers_and_usaplayers_each_year['name']/total_players_each_year['name']

x_axis = np.arange(2007,2018,1)
y_axis_usa = percentage_each_year.groupby(level='year').tail(1)
y_axis_international = percentage_each_year.groupby(level='year').head(1)
```

```python
#plotting
plt.plot(x_axis,y_axis_usa,label='USA',marker='>',c="#552583")
plt.plot(x_axis,y_axis_international,label='International',marker=">",c = "#FDB927")
plt.title("The Trend of NBA Player Composition of the Years from 2007 - 2017")
plt.xlabel("Year")
plt.ylabel("% of Two Categories")
plt.legend()
plt.grid()
```

# Heat Layer of Salary

```python
#remove the $ and commas
unique_name_df[' Salary in $ '] = unique_name_df[' Salary in $ '].str.replace(',','').str.replace('$','').astype('float')
```

```python
unique_name_df.head()
```

```python
import gmaps
# Configure gmaps with API key
gmaps.configure(api_key=gkey)

#add heat layer
fig = gmaps.figure()

unique_locations = unique_name_df[['Lat','Lng']]


salary_list = unique_name_df[' Salary in $ '].tolist()

heat_layer = gmaps.heatmap_layer(unique_locations, weights=salary_list, 
                                 dissipating=False, max_intensity=19000000,
                                 point_radius = 1)

fig.add_layer(heat_layer)

fig
```
